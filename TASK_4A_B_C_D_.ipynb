```python
import numpy as np
import matplotlib.pyplot as plt
x=data['INDEPEND']
y=data['CSFVOL']
labels=data['CDRGLOB']
plt.scatter(x,y, cmap='viridis',c=labels, alpha=0.5)
plt.xlabel('ındependent living')
plt.ylabel('Cerebral Spinal Fluid Volume')
plt.title('Independ-CSFVOL')
plt.colorbar()
plt.show()
```

TASK-4D:Please provide a several pivot table for insightful illustration both via tables and figures.

```python
import pandas as pd 
rsumfrm=pd.qcut(data['RSUPFRM'],4)
csfvol=pd.qcut(data['CSFVOL'],6)
data.pivot_table('HIPPOVOL',index=['CDRGLOB',csfvol],columns=[rsumfrm])
```

```python
import pandas as pd 
rsumfrm2=pd.qcut(data['RSUPFRM'],2)
data.pivot_table('CSFVOL',index=['CDRGLOB','SEX'],columns=[rsumfrm2]).plot(kind='bar')
plt.legend()
```
